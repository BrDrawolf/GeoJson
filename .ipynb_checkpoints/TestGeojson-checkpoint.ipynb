```python
import pandas as pd
import numpy as np
from pandas.io.json import json_normalize
import json
import utm
```

```python
with open('MACROBUS.geojson') as f:
    geojson = json.load(f)
```

```python
df = json_normalize(geojson['features'])
```

```python
df.dtypes
```

```python
df
```

```python
df.iloc[0]['geometry.coordinates'][0]
```

```python
def getlatlons(row):
    return pd.Series(utm.to_latlon(row[0], row[1], 13, 'Q'))
```

```python
newCoordinates = df[['properties.X','properties.Y']].apply(getlatlons,axis=1)
newCoordinates.columns = ['lat', 'lon']

newCoordinates
```

```python
newCoordinates.to_csv('Macrobus.csv')
```
